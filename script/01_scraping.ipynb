```python
# Cell 1: Install & Import Dependencies
!pip install google-play-scraper pandas
import pandas as pd
from google_play_scraper import reviews, Sort
import json

print("Dependencies installed and imported.")
```

```python
# Cell 2: Define Scraping Parameters
# ID Aplikasi Tokopedia di Google Play Store
APP_ID = 'com.tokopedia.tkpd'
TARGET_REVIEWS = 10000  # Target: 10.000+ data
LANGUAGE = 'id'
COUNTRY = 'id'

print(f"Start scraping for {APP_ID}...")
```

```python
# Cell 3: Run the Scraper
# Kita akan menggunakan 'reviews' dalam loop dengan continuation_token untuk data besar
# 'reviews_all' terkadang tidak stabil untuk data > 10.000

all_reviews = []
continuation_token = None

while len(all_reviews) < TARGET_REVIEWS:
    try:
        # Ambil 200 review sekaligus
        result, continuation_token = reviews(
            APP_ID,
            lang=LANGUAGE,
            country=COUNTRY,
            sort=Sort.NEWEST,  # Ambil yang terbaru
            count=200,
            continuation_token=continuation_token
        )
        
        if not result:
            print("Tidak ada review lagi.")
            break
            
        all_reviews.extend(result)
        
        if continuation_token is None:
            print("Scraping selesai, tidak ada continuation_token.")
            break
            
        print(f"Total reviews scraped: {len(all_reviews)}")

    except Exception as e:
        print(f"Terjadi error: {e}")
        break

print(f"Scraping selesai. Total {len(all_reviews)} reviews didapatkan.")
```

```python
# Cell 4: Convert to DataFrame and Save
df = pd.DataFrame(all_reviews)

# Pilih kolom yang relevan
df_final = df[['reviewId', 'userName', 'content', 'score', 'at']]

print(f"Total data: {len(df_final)}")
print(df_final.head())
```

```python
# Cell 5: Save to CSV (Deliverable: dataset.csv)
# Pastikan folder data/raw sudah ada
import os
output_dir = '../data/raw'
output_file = f'{output_dir}/playstore_reviews.csv'

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

df_final.to_csv(output_file, index=False)
print(f"Data berhasil disimpan di {output_file}")
```
