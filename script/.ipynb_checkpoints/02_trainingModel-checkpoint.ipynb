```python
# Import Libraries
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline as ImbPipeline 

# Scikit-learn
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
import joblib

# Transformers (Deep Learning)
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from datasets import Dataset
from sklearn.metrics import precision_recall_fscore_support, accuracy_score

# Setup
pd.set_option('display.max_colwidth', None)
sns.set_style('whitegrid')

# NLTK stopwords
nltk.download('stopwords')
```

```python
# Cell 2: Load Data
df = pd.read_csv('../data/raw/playstore_reviews.csv')
df.info()
df.head()
```

```python
# Cell 3: Phase 2 - EDA & Preprocessing
print("--- EDA & Preprocessing ---")

# 1. Handle Missing Values
print(f"Data sebelum drop NA: {len(df)}")
df = df.dropna(subset=['content'])
print(f"Data setelah drop NA: {len(df)}")

# 2. Handle Duplicates
df = df.drop_duplicates(subset=['content'])
print(f"Data setelah drop duplikat: {len(df)}")

# 3. Labeling
def map_sentiment(score):
    if score <= 3:        # Bintang 1, 2, dan 3 adalah 'negatif'
        return 'negatif'
    else:                 # Bintang 4 dan 5 adalah 'positif'
        return 'positif'

df['sentiment'] = df['score'].apply(map_sentiment)

# 4. Visualisasi Distribusi Kelas
print("\nDistribusi Kelas:")
print(df['sentiment'].value_counts())
sns.countplot(x='sentiment', data=df, order=['positif', 'negatif'])
plt.title('Distribusi Sentimen Ulasan')
plt.show()
```

```python
# Cell 4: Text Cleaning
print("--- Text Cleaning ---")
# Ambil stopword bahasa Indonesia
list_stopwords = stopwords.words('indonesian')

# Definisikan kata-kata negasi yang ingin kita PERTAHANKAN
negation_words = [
    'tidak', 'bukan', 'jangan', 'bkn', 'tdk', 'ga', 'gk', 'enggak'
]

# Buat list stopword baru TANPA kata-kata negasi
list_stopwords = [word for word in list_stopwords if word not in negation_words]

def clean_text(text):
    text = text.lower() # Lowercasing
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE) # Hapus URLs
    text = re.sub(r'\@\w+|\#','', text) # Hapus mentions dan hashtags
    text = re.sub(r'[^\w\s]', '', text) # Hapus Punctuation
    text = re.sub(r'\d+', '', text) # Hapus angka
    text = re.sub(r'\s+', ' ', text).strip() # Hapus whitespace berlebih
    return text

def remove_stopwords(text):
    tokens = text.split()
    cleaned_tokens = [word for word in tokens if word not in list_stopwords]
    return ' '.join(cleaned_tokens)

# Buat 2 versi teks:
# 1. basic_cleaned: Hanya cleaning dasar (untuk BERT)
df['text_cleaned'] = df['content'].apply(clean_text)

# 2. preprocessed: Cleaning + Stopword removal (untuk TF-IDF)
df['text_preprocessed'] = df['text_cleaned'].apply(remove_stopwords)

# Tampilkan perbandingan
print(df[['content', 'text_cleaned', 'text_preprocessed']].head())

# Simpan data yang sudah diproses
df.to_csv('../data/processed/cleaned_reviews.csv', index=False)
```

```python
# Cell 5: Word Cloud Visualization
print("--- Word Cloud (Positif vs Negatif) ---")
plt.figure(figsize=(15, 7))

# Word Cloud Positif
text_positif = " ".join(review for review in df[df['sentiment'] == 'positif']['text_preprocessed'])
wordcloud_positif = WordCloud(max_words=100, background_color="white").generate(text_positif)
plt.subplot(1, 2, 1)
plt.imshow(wordcloud_positif, interpolation='bilinear')
plt.title('Word Cloud - Sentimen Positif')
plt.axis("off")

# Word Cloud Negatif
text_negatif = " ".join(review for review in df[df['sentiment'] == 'negatif']['text_preprocessed'])
wordcloud_negatif = WordCloud(max_words=100, background_color="black").generate(text_negatif)
plt.subplot(1, 2, 2)
plt.imshow(wordcloud_negatif, interpolation='bilinear')
plt.title('Word Cloud - Sentimen Negatif')
plt.axis("off")

plt.show()
```

```python
# Cell 6: Data Splitting (Train 80%, Test 20%)
# Map label ke integer
label_map = {'negatif': 0, 'positif': 1}
df['label'] = df['sentiment'].map(label_map)

# Split data
# X_ml -> text_preprocessed (untuk TF-IDF)
# X_dl -> text_cleaned (untuk BERT)
# y -> label
X_ml = df['text_preprocessed']
X_dl = df['text_cleaned']
y = df['label']

X_train_ml, X_test_ml, y_train, y_test = train_test_split(X_ml, y, test_size=0.2, random_state=42, stratify=y)
X_train_dl, X_test_dl = train_test_split(X_dl, test_size=0.2, random_state=42, stratify=y)

print(f"Train data size: {len(X_train_ml)}")
print(f"Test data size: {len(X_test_ml)}")
```

```python
# Cell 7: ================ SKEMA 1 ================
# Algoritma: Logistic Regression
# Ekstraksi Fitur: TF-IDF
# Pembagian Data: 80/20

print("\n--- SKEMA 1: Logistic Regression + TF-IDF ---")

# Buat pipeline
pipeline_logreg = Pipeline([
    ('tfidf', TfidfVectorizer(ngram_range=(1, 2))),
    ('clf', LogisticRegression(random_state=42, max_iter=1000,))
])

# Training
pipeline_logreg.fit(X_train_ml, y_train)

# Prediksi
y_pred_logreg = pipeline_logreg.predict(X_test_ml)

# Evaluasi
print("Hasil Evaluasi Skema 1 (Logistic Regression):")
print(classification_report(y_test, y_pred_logreg, target_names=label_map.keys()))
print(f"Akurasi Test: {accuracy_score(y_test, y_pred_logreg) * 100:.2f}%")
# Akurasi testing HARUS >= 85%
```

```python
# Cell 8: ================ SKEMA 2 ================
# Algoritma: SVM (Support Vector Machine)
# Ekstraksi Fitur: TF-IDF
# Pembagian Data: 80/20

print("\n--- SKEMA 2: SVM + TF-IDF ---")

# Buat pipeline
pipeline_svm = Pipeline([
    ('tfidf', TfidfVectorizer(ngram_range=(1, 2))),
    ('clf', SVC(kernel='linear', random_state=42,))
])

# Training
pipeline_svm.fit(X_train_ml, y_train)

# Prediksi
y_pred_svm = pipeline_svm.predict(X_test_ml)

# Evaluasi
print("Hasil Evaluasi Skema 2 (SVM):")
print(classification_report(y_test, y_pred_svm, target_names=label_map.keys()))
print(f"Akurasi Test: {accuracy_score(y_test, y_pred_svm) * 100:.2f}%")
# Akurasi testing HARUS >= 85%

# Simpan model ML terbaik (misal: SVM jika lebih baik)
joblib.dump(pipeline_svm, '../models/ml_pipeline.pkl')
print("Model SVM (Skema 2) disimpan ke models/ml_pipeline.pkl")
```

```python
# Cell 9: ================ SKEMA 3 ================
# Algoritma: Multinomial Naive Bayes
# Ekstraksi Fitur: TF-IDF
# Sampling: Random Undersampling

print("\n--- SKEMA 3: Naive Bayes + TF-IDF + Undersampling ---")

# pipeline MENGGUNAKAN ImbPipeline
# sampling HANYA pada data training, tidak pada data testing
pipeline_nb = ImbPipeline([
    ('tfidf', TfidfVectorizer(ngram_range=(1, 2), max_features=20000)),
    ('sampling', RandomUnderSampler(random_state=42)), 
    ('clf', MultinomialNB())
])

# Training
pipeline_nb.fit(X_train_ml, y_train)

# Prediksi
y_pred_nb = pipeline_nb.predict(X_test_ml)

# Evaluasi
print("Hasil Evaluasi Skema 3 (Naive Bayes + Undersampling):")
print(classification_report(y_test, y_pred_nb, target_names=label_map.keys()))
print(f"Akurasi Test: {accuracy_score(y_test, y_pred_nb) * 100:.2f}%")
```

```python
# Cell 15: ================ Perbandingan Skema ================
print("\n--- PERBANDINGAN AKURASI TEST SET ---")
print(f"Skema 1 (LogReg + TF-IDF): {accuracy_score(y_test, y_pred_logreg) * 100:.2f}%")
print(f"Skema 2 (SVM + TF-IDF):    {accuracy_score(y_test, y_pred_svm) * 100:.2f}%")
print(f"Skema 3 (NB + TF-IDF):     {accuracy_score(y_test, y_pred_nb) * 100:.2f}%")
```
